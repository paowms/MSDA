```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
df = pd.read_csv(r'C:\Users\paowm\Downloads\teleco_market_basket.csv')
df.info()
```

```python
df.shape
```

```python
df.head()
```

```python
df.shape
```

```python
tr = [] 
for i in range (0,7500):
    tr.append([str(df.values[i,j]) for j in range(0, 20)])
```

```python
te = TransactionEncoder()
array = te.fit(tr).transform(tr)
```

```python
clean_df = pd.DataFrame(array, columns = te.columns_)
clean_df
```

```python
for col in clean_df.columns:
    print(col)
```

```python
df_clean = clean_df.drop(['nan'], axis = 1)
df_clean
```

```python
for col in df_clean.columns:
    print(col)
```

```python
df_clean.shape
```

```python
df_clean.to_csv(r'C:\Users\paowm\Downloads\market_basket_clean.csv')
```

```python
rules = apriori(df_clean, min_support = 0.025, use_colnames = True)
rules.head()
```

```python
assoc_rules = association_rules(rules, metric = 'lift', min_threshold = 1)
assoc_rules.head()
```

```python
assoc_rules.sort_values(by = ["lift"],ascending=False)
```
