```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import statistics as st
import numpy as np
import seaborn as sns

##if not np.version.version[0:4] == "1.23":
  ##  raise f'not {np.version.version[0:4]=}!=1.23'
```

```python
df = pd.read_csv(r"C:\Users\paowm\Downloads\d9rkejv84kd9rk30fi2l\churn_clean.csv")
```

```python
x = df[df["Churn"]=="Yes"]["Outage_sec_perweek"]
y = df[df["Churn"]=="No"]["Outage_sec_perweek"]
```

```python
#Prove Normality of datasets
sns.displot(data=x, kind="hist", kde=True)
sns.displot(data=y, kind="hist", kde=True)
```

```python
t_result = stats.ttest_ind(x,y)

alpha= 0.05
print("The alpha/2 value is 0.025")
print("The p-value is " + str(t_result[1]))
if(t_result[1] < alpha):
    print("The means of x and y are different")
else:
	print("No significant difference found")
```
